# 영화관 좌석 티켓 계산
[A,B][1:12] = EC = 7,000원 [C:F][1:2],[11,12] = EC / [C:F][3:10] = RE = 8,000원 [G:H][1:12] = SP = 10,000원

```python
Seats={
    'A':['EC','EC','EC','EC','EC','EC','EC','EC','EC','EC','EC','EC'], #A - 0
    'B':['EC','EC','EC','EC','EC','EC','EC','EC','EC','EC','EC','EC'], #B - 1
    'C':['EC','EC','RE','RE','RE','RE','RE','RE','RE','RE','EC','EC'], #C - 2
    'D':['EC','EC','RE','RE','RE','RE','RE','RE','RE','RE','EC','EC'], #D - 3
    'E':['EC','EC','RE','RE','RE','RE','RE','RE','RE','RE','EC','EC'], #E - 4
    'F':['EC','EC','RE','RE','RE','RE','RE','RE','RE','RE','EC','EC'], #F - 5
    'G':['SP','SP','SP','SP','SP','SP','SP','SP','SP','SP','SP','SP'], #G - 6
    'H':['SP','SP','SP','SP','SP','SP','SP','SP','SP','SP','SP','SP']  #H - 7
}



def ShowSeats(T):
    #print(T)
    #형식에 맞게 출력
    print('- 좌석배치도 :\n')
    strLine = ''
    for n in range(1,12+1,1):
        if n>=10 : strLine += '   |  '+str(n)
        else:strLine += '   |   '+str(n)
    strLine += '   |   '
    print(strLine)#1 2 3... 12
    print('-'*110)
    
    strLine=''
    for key in Seats:
        strLine=''
        strLine+=key
        strLine += '  |   '
        for value in Seats[key]:
            strLine+=value
            strLine += '  |   '
        #strLine+='\n'    
        print(strLine)
        print('-'*110)
    
    
def Reserve(T, key, col):
    T[key][col]='**'
    

#Seats[0][1] = A열의 1번째 자리
ECPrice = 7000
REPrice = 8000
SPPrice = 10000
print('EC : {}원, RE : {}원, SP : {}원'.format(ECPrice,REPrice, SPPrice))

while True :
    ShowSeats(Seats)
    SelectSeats = input('원하는 좌석을 선택해주세요(예:F12, 종료:X) : ')
    if SelectSeats.upper() == 'X':
        print('예약이 종료되었습니다.')
        break
        
    alpa = SelectSeats[0]
    num = int(SelectSeats[1:3])-1
    
    if not(alpa == 'A'
       or alpa == 'B'
       or alpa == 'C'
       or alpa == 'D'
       or alpa == 'E'
       or alpa == 'F'
       or alpa == 'G'
       or alpa == 'H'):
        print('좌석번호가 잘못되었습니다. 다시 입력해주세요!')
        
    elif(num > 12):
        print('좌석번호가 잘못되었습니다. 다시 입력해주세요!')
        
    else:
        print('좌석번호 : {}'.format(SelectSeats))
        if(Seats[alpa][num] == 'EC'):
            print('티켓가격 : {}원'.format(ECPrice))
            
        elif(Seats[alpa][num] == 'RE'):
            print('티켓가격 : {}원'.format(REPrice))
            
        elif(Seats[alpa][num] == 'SP'):
            print('티켓가격 : {}원'.format(SPPrice))

        #break
        
    YesNo = input('예약하시겠습니까? Y or N : ')
    if YesNo.upper() == 'Y':
         Seats[alpa][num]='**'
    elif YesNo.upper() == 'N':
        print('예약이 종료되었습니다.')
        break
        
        
```
