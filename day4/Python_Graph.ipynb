# 3차 함수 : f(x) = (x-2)*x*(x+2)

```python
import matplotlib.pyplot as plt

def f(x):
    y = (x-2)*x*(x+2)
    return y

print('x = 1', f(1))
print('x = 2', f(2))
print('x = 3', f(3))

x = np.array([1,2,3])#list -> np.ndarray
x = np.arange(-3,3.5,0.5)#np.arange -> 범위 지정
x = np.linspace(-3,3,50) #start, finish, samples
print('x [] = {} type = {}'.format(x, type(x)))
print(f(x))


plt.figure(figsize=(5,5))
plt.plot(x,f(x),'cornflowerblue', linewidth = 3, label = '$y=(x-2)*x*(x+2)$')
plt.ylim(-30,30)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='lower right')
plt.show()
plt.savefig('graph.png')
```

```python
import matplotlib.pyplot as plt

def f(x):
    y = (x-2)*x*(x+2)
    return y

def f2(x,w):
    y = (x-w)*x*(x+2)
    return y

print('x = 1', f(1))
print('x = 2', f(2))
print('x = 3', f(3))

x = np.array([1,2,3])#list -> np.ndarray
x = np.arange(-3,3.5,0.5)#np.arange -> 범위 지정
x = np.linspace(-3,3,50) #start, finish, samples
print('x [] = {} type = {}'.format(x, type(x)))
print(f(x))


plt.figure(figsize=(5,5))
#plt.plot(x,f(x),'cornflowerblue', linewidth = 3, label = '$y=(x-2)*x*(x+2)$')
plt.plot(x,f(x),'cornflowerblue', linewidth = 3) # 상자에 f(x)안나타남
plt.plot(x,f2(x,w=0),'black', linewidth = 3, label = '$y=(x-0)*x*(x+2)$')
plt.plot(x,f2(x,w=4),'gray', linewidth = 3, label = '$w=4$')
plt.plot(x,x,'orange', linewidth = 3, label = '$y=x$')

plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')

plt.ylim(-30,30)
plt.xlim(-4,4)
plt.grid(True)
plt.legend(loc='upper left')
plt.show()
plt.savefig('graph.png')
```

```python
def f3(x0, x1):
    r = 2*x0**2 + x1**2
    ans = r*np.exp(-r)
    return ans

#x0, x1에 따라서 각각 f3함수를 계산
xn = 9
x0 = np.linspace(-2,2, xn)
x1 = np.linspace(-2,2, xn)
y = np.zeros((len(x0),len(x1)))
#print('x0 = {}'.format(x0))
#print('x1 = {}'.format(x1))
#print('y = {}'.format(y))

for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
        
y = np.round(y,1)
#print('y = {}'.format(y))
plt.figure(figsize = (3.5,3))
#plt.gray()
plt.bone()
plt.pcolor(y)
plt.colorbar()
#plt.show()
plt.savefig('gauss.png')
```

```python
from mpl_toolkits.mplot3d import Axes3D
xn = 9
x0 = np.linspace(-2,2, xn)
x1 = np.linspace(-2,2, xn)
#print(x0)
#print(x1)
xx0, xx1 = np.meshgrid(x0, x1)

#print(xx0)
#print(xx1)
plt.figure(figsize = (5,3.5))
ax = plt.subplot(1,1,1, projection='3d')
ax.plot_surface(xx0, xx1, y, rstride = 1, cstride = 1, alpha = 0.3,
               color = 'blue', edgecolor = 'black')
ax.set_zticks((0,0.2))
ax.view_init(40, -115)
```

# 그래프 표시 - contour

```python
xn = 50
x0 = np.linspace(-2,2, xn)
x1 = np.linspace(-2,2, xn)
y = np.zeros((len(x0),len(x1)))

for row in range(xn):
    for col in range(xn):
        y[col, row] = f3(x0[row], x1[col])
xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(figsize = (4,4))
cont = plt.contour(xx0, xx1, y, 5, colors= 'black')
cont.clabel(fmt='%3.2f', fontsize = 8)
plt.xlabel('$x_0$', fontsize = 14)
plt.ylabel('$x_1$', fontsize = 14)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,3))
plt.subplots_adjust(wspace=0.5, hspace=0.5)

xn = 50
x0 = np.linspace(-2,2, xn)
x1 = np.linspace(-2,2, xn)
xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(figsize = (6,4.5))

for i in range(4):
   # plt.subplot(2,3,i+1) #plot 묘사의 위치 지정
    ax = plt.subplot(2,2,i+1, projection='3d')
    ax.plot_surface(xx0, xx1, y, rstride = 1, cstride = 1, alpha = 0.5, color = 'blue', edgecolor = 'black')
    ax.set_zticks((0,0.2))
    plt.title(i+1)
    if i==0:
        ax.view_init(40,0)
    elif i==1:
        ax.view_init(40,0)
    elif i==2:
        ax.view_init(1, -1)
    elif i==3:
        ax.view_init(45, -115)
        
plt.show()
```

# Quiz. 구구단 3단, 5단, 7단, 9단을 Plot(1개), SubPlot(4분할)로 그래프 표현

```python
import matplotlib.pyplot as plt
def GuGu(dan): #ndarray
    x = np.linspace(1,9, 9)
    y = dan*x
    return y
print(x)
print(y)
print(dan)
x = np.linspace(1,9,9)
plt.figure(figsize = (5,5))
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)

subPlotIndex = 1
for i in range(3,10,2):
    plt.subplot(2,2,subPlotIndex)
    titleName = '{}Dan'.format(i)
    plt.title(titleName)
    plt.plot(x,GuGu(i), 'k')
    plt.ylim(0, 91)
    plt.grid(True)
    subPlotIndex += 1
plt.show()

plt.figure(figsize=(5,5))
plt.plot(x,GuGu(3),'cornflowerblue', linewidth = 3, label = '$y=3*x$')
plt.plot(x,GuGu(5),'black', linewidth = 3, label = '$y=5*x$')
plt.plot(x,GuGu(7),'yellow', linewidth = 3, label = '$y=7*x$')
plt.plot(x,GuGu(9),'green', linewidth = 3, label = '$y=9*x$')
```
